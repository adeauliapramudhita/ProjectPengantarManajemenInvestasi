## 2. Metode yang Digunakan

Metode yang digunakan adalah **Modern Portfolio Theory (MPT)** oleh Harry Markowitz, khususnya analisis **Efficient Frontier**. Tujuannya untuk menentukan kombinasi portofolio optimal berdasarkan hubungan antara **return** dan **risiko (volatilitas)**.

### 2.1 Perhitungan Return Harian

Return harian dihitung dari perubahan harga penutupan sebagai:

$$
r_{i,t} = \frac{P_{i,t} - P_{i,t-1}}{P_{i,t-1}}
$$

Keterangan:

* $r_{i,t}$ = return harian saham ke-$i$ pada hari ke-$t$
* $P_{i,t}$ = harga penutupan saham ke-$i$ pada hari ke-$t$
* $P_{i,t-1}$ = harga penutupan saham ke-$i$ pada hari sebelumnya

---

### 2.2 Return Ekspektasi Tahunan

Return ekspektasi tahunan diperoleh dari rata-rata return harian dengan asumsi **252 hari bursa per tahun**:

$$
\mu_i = \bar{r}_{i,\text{daily}} \times 252
$$

Keterangan:

* $\mu_i$ = return ekspektasi tahunan saham ke-$i$
* $\bar{r}_{i,\text{daily}}$ = rata-rata return harian saham ke-$i$

---

### 2.3 Risiko (Volatilitas) Portofolio

Risiko portofolio dihitung menggunakan **varians-kovarians** antar saham:

$$
\sigma_p = \sqrt{ \mathbf{w}^\top \Sigma \mathbf{w} }
$$

Keterangan:

* $\sigma_p$ = volatilitas (risiko) portofolio
* $\mathbf{w}$ = vektor bobot saham dalam portofolio
* $\Sigma$ = matriks kovarians return tahunan antar saham

Matriks kovarians diperoleh dengan:

$$
\Sigma = \text{Cov}(r_i, r_j) \times 252
$$

---

### 2.4 Return Portofolio

Return portofolio adalah kombinasi tertimbang dari return ekspektasi tiap saham:

$$
R_p = \mathbf{w}^\top \mu
$$

Keterangan:

* $R_p$ = return ekspektasi portofolio
* $\mu$ = vektor return ekspektasi tahunan tiap saham
* $\mathbf{w}$ = bobot saham dalam portofolio ($\sum_i w_i = 1$)

---

### 2.5 Sharpe Ratio

Untuk mengukur kinerja portofolio relatif terhadap risiko digunakan **Sharpe Ratio**:

$$
S = \frac{R_p - R_f}{\sigma_p}
$$

Keterangan:

* $S$ = Sharpe Ratio
* $R_p$ = return ekspektasi portofolio
* $R_f$ = risk-free rate (diasumsikan 5%)
* $\sigma_p$ = volatilitas portofolio

Portofolio dengan nilai $S$ terbesar disebut sebagai **portofolio optimal** (Max Sharpe Portfolio).

---

### 2.6 Efficient Frontier

Dengan melakukan simulasi Monte Carlo (50.000 kali kombinasi bobot acak), diperoleh sekumpulan titik $(\sigma_p, R_p)$. Dari kumpulan ini:

* Bagian “terluar” menggambarkan **Efficient Frontier**, yaitu portofolio dengan return tertinggi untuk tingkat risiko tertentu.
* Portofolio dengan risiko terkecil disebut **Minimum Variance Portfolio**.

---

Dengan data harga saham HIMBARA + BSI dan metode Markowitz, diperoleh dua portofolio penting:

1. **Portofolio Optimal (Max Sharpe)** → return relatif tinggi dengan risiko efisien.
2. **Portofolio Minimum Variance** → risiko terkecil walaupun return lebih rendah.

## **Max Sharpe**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Jumlah simulasi portofolio
num_portfolios = 50000
risk_free_rate = 0.05 # misalnya BI rate 5%

results = np.zeros((4, num_portfolios))
weights_record = []

for i in range(num_portfolios):
    # Random bobot saham
    weights = np.random.random(len(mean_return))
    weights /= np.sum(weights)
    weights_record.append(weights)

    # Hitung return & risiko portofolio
    port_return = np.dot(weights, mean_return)
    port_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    sharpe_ratio = (port_return - risk_free_rate) / port_volatility

    results[0,i] = port_return
    results[1,i] = port_volatility
    results[2,i] = sharpe_ratio
    results[3,i] = i

# Cari portofolio optimal
max_sharpe_idx = np.argmax(results[2,:])  # Portofolio dengan Sharpe ratio tertinggi

# Plot Efficient Frontier
plt.figure(figsize=(12,8))
scatter = plt.scatter(results[1,:], results[0,:], c=results[2,:], cmap='viridis', marker='o', s=10)
plt.colorbar(label='Sharpe Ratio')

# Tambahkan titik bintang untuk portofolio optimal
plt.scatter(results[1, max_sharpe_idx], results[0, max_sharpe_idx],
            marker='*', color='red', s=400, edgecolors='black',
            label=f'Max Sharpe Ratio ({results[2, max_sharpe_idx]:.3f})')

plt.xlabel('Risiko (Volatilitas)')
plt.ylabel('Return Ekspektasi')
plt.title('Efficient Frontier Saham HIMBARA dengan Portofolio Optimal')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"\n1. PORTOFOLIO MAX SHARPE RATIO:")
print(f"   Sharpe Ratio: {results[2, max_sharpe_idx]:.4f}")
print(f"   Return: {results[0, max_sharpe_idx]:.4f}")
print(f"   Volatilitas: {results[1, max_sharpe_idx]:.4f}")
print("   Bobot Saham:")
max_sharpe_weights = weights_record[max_sharpe_idx]
for stock, weight in zip(mean_return.index, max_sharpe_weights):
    if weight > 0.0001:
        print(f"     {stock}: {weight:.4f}")
```

## **Minimum Variance**

### **Minimum Varians 1 Constraint**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

risk_free_rate = 0.05

# =============================
# PORTFOLIO MINIMUM VARIANCE (1 CONSTRAINT)
# =============================
def min_variance_portfolio(cov_matrix):
    """Portofolio dengan varians minimum, constraint: sum(weights) = 1"""
    n_assets = len(cov_matrix)

    # Fungsi objective: minimize variance
    def objective(weights):
        return weights @ cov_matrix @ weights

    # Constraints: sum(weights) = 1
    constraints = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1})

    # Bounds: weights between 0 and 1
    bounds = tuple((0, 1) for _ in range(n_assets))

    # Initial guess: equal weights
    initial_weights = np.ones(n_assets) / n_assets

    # Optimize
    result = minimize(objective, initial_weights,
                     method='SLSQP', bounds=bounds, constraints=constraints)

    return result.x

# Eksekusi Minimum Variance Portfolio
print("="*60)
print("PORTFOLIO MINIMUM VARIANCE (1 CONSTRAINT)")
print("="*60)

weights_min_var = min_variance_portfolio(cov_matrix.values)
return_min_var = np.dot(weights_min_var, mean_return.values)
vol_min_var = np.sqrt(np.dot(weights_min_var.T, np.dot(cov_matrix.values, weights_min_var)))
sharpe_min_var = (return_min_var - risk_free_rate) / vol_min_var

print(f"Return: {return_min_var:.6f}")
print(f"Volatilitas: {vol_min_var:.6f}")
print(f"Sharpe Ratio: {sharpe_min_var:.6f}")
print("Bobot Saham:")
for stock, weight in zip(mean_return.index, weights_min_var):
    if weight >= 0.0000:
        print(f"  {stock}: {weight:.6f}")

# =============================
# SIMULASI EFFICIENT FRONTIER (MONTE CARLO)
# =============================
print("\n" + "="*60)
print("GENERATING EFFICIENT FRONTIER")
print("="*60)

num_portfolios = 50000
results = np.zeros((3, num_portfolios))
weights_record = []

print(f"Menghasilkan {num_portfolios} portofolio acak...")

for i in range(num_portfolios):
    # Random bobot saham
    weights = np.random.random(len(mean_return))
    weights /= np.sum(weights)
    weights_record.append(weights)

    # Hitung return & risiko portofolio
    port_return = np.dot(weights, mean_return.values)
    port_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix.values, weights)))
    sharpe_ratio = (port_return - risk_free_rate) / port_volatility

    results[0,i] = port_return
    results[1,i] = port_volatility
    results[2,i] = sharpe_ratio

# =============================
# VISUALISASI EFFICIENT FRONTIER
# =============================
plt.figure(figsize=(12, 8))

# Plot efficient frontier (semua portofolio acak)
scatter = plt.scatter(results[1,:], results[0,:], c=results[2,:], cmap='viridis',
                     marker='o', s=10)
plt.colorbar(label='Sharpe Ratio')

# Plot portofolio minimum variance
plt.scatter(vol_min_var, return_min_var, marker='*', color='red', s=400,
           edgecolors='black', linewidth=2, label='Minimum Variance Portfolio')

# Konfigurasi plot
plt.xlabel('Volatilitas (Risiko)', fontsize=12)
plt.ylabel('Return Ekspektasi', fontsize=12)
plt.title('Efficient Frontier dengan Portofolio Optimal', fontsize=14, fontweight='bold')
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)

# Tambahkan anotasi untuk portofolio optimal
plt.annotate(f'Min Variance\n({return_min_var:.3f}, {vol_min_var:.3f})',
            xy=(vol_min_var, return_min_var),
            xytext=(vol_min_var + 0.005, return_min_var - 0.001),
            fontsize=10, ha='left')

plt.tight_layout()
plt.show()
```

### **Maximum Return**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

risk_free_rate = 0.05

# =============================
# PORTFOLIO MAXIMUM RETURN (1 CONSTRAINT)
# =============================
def max_return_portfolio(mean_return):
    """Portofolio dengan return maksimum, constraint: sum(weights) = 1"""
    n_assets = len(mean_return)

    # Fungsi objective: maximize return -> minimize negative return
    def objective(weights):
        return -weights @ mean_return

    # Constraints: sum(weights) = 1
    constraints = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1})

    # Bounds: weights between 0 and 1
    bounds = tuple((0, 1) for _ in range(n_assets))

    # Initial guess
    initial_weights = np.ones(n_assets) / n_assets

    # Optimize
    result = minimize(objective, initial_weights,
                     method='SLSQP', bounds=bounds, constraints=constraints)

    return result.x

# Eksekusi Maximum Return Portfolio
print("="*60)
print("PORTFOLIO MAXIMUM RETURN (1 CONSTRAINT)")
print("="*60)

weights_max_return = max_return_portfolio(mean_return.values)
return_max_return = np.dot(weights_max_return, mean_return.values)
vol_max_return = np.sqrt(np.dot(weights_max_return.T, np.dot(cov_matrix.values, weights_max_return)))
sharpe_max_return = (return_max_return - risk_free_rate) / vol_max_return

print(f"Return: {return_max_return:.6f}")
print(f"Volatilitas: {vol_max_return:.6f}")
print(f"Sharpe Ratio: {sharpe_max_return:.6f}")
print("Bobot Saham:")
for stock, weight in zip(mean_return.index, weights_max_return):
    if weight > 0.001:
        print(f"  {stock}: {weight:.6f}")

# =============================
# SIMULASI EFFICIENT FRONTIER (MONTE CARLO)
# =============================
print("\n" + "="*60)
print("GENERATING EFFICIENT FRONTIER")
print("="*60)

num_portfolios = 50000
results = np.zeros((3, num_portfolios))
weights_record = []

print(f"Menghasilkan {num_portfolios} portofolio acak...")

for i in range(num_portfolios):
    # Random bobot saham
    weights = np.random.random(len(mean_return))
    weights /= np.sum(weights)
    weights_record.append(weights)

    # Hitung return & risiko portofolio
    port_return = np.dot(weights, mean_return.values)
    port_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix.values, weights)))
    sharpe_ratio = (port_return - risk_free_rate) / port_volatility

    results[0,i] = port_return
    results[1,i] = port_volatility
    results[2,i] = sharpe_ratio

# =============================
# VISUALISASI EFFICIENT FRONTIER
# =============================
plt.figure(figsize=(12, 8))

# Plot efficient frontier (semua portofolio acak)
scatter = plt.scatter(results[1,:], results[0,:], c=results[2,:], cmap='viridis',
                     marker='o', s=10)
plt.colorbar(label='Sharpe Ratio')

# Plot portofolio maximum return
plt.scatter(vol_max_return, return_max_return, marker='*', color='red', s=400,
           edgecolors='black', linewidth=2, label='Maximum Return Portfolio')

# Konfigurasi plot
plt.xlabel('Volatilitas (Risiko)', fontsize=12)
plt.ylabel('Return Ekspektasi', fontsize=12)
plt.title('Efficient Frontier dengan Portofolio Maximum Return', fontsize=14, fontweight='bold')
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)

# Tambahkan anotasi untuk portofolio optimal
plt.annotate(f'Max Return\n({return_max_return:.3f}, {vol_max_return:.3f})',
            xy=(vol_max_return, return_max_return),
            xytext=(vol_max_return + 0.005, return_max_return - 0.001),
            fontsize=10, ha='left')

plt.tight_layout()
plt.show()
```

## **Firefly Algorithm**


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter Firefly Algorithm
n_fireflies = 30       # jumlah firefly
max_iter = 200         # jumlah iterasi maksimum
alpha = 0.2            # randomization parameter
beta0 = 1              # attractiveness at r=0
gamma = 1              # light absorption coefficient

n_assets = len(mean_return)
r_f = 0.05

# Fungsi objektif: maximize Sharpe Ratio
def fitness(w, mean_return, cov_matrix, r_f=0.0):
    port_return = np.dot(w, mean_return)
    port_vol = np.sqrt(np.dot(w.T, np.dot(cov_matrix, w)))
    sharpe = (port_return - r_f) / port_vol
    return -sharpe  # negatif karena FA minimisasi

# Normalisasi bobot agar sum=1 dan >=0
def normalize(w):
    w = np.clip(w, 0, 1)
    return w / np.sum(w)

# Inisialisasi populasi
fireflies = np.random.dirichlet(np.ones(n_assets), size=n_fireflies)
fitness_vals = np.array([fitness(w, mean_return, cov_matrix, r_f) for w in fireflies])

# Iterasi Firefly Algorithm
for t in range(max_iter):
    for i in range(n_fireflies):
        for j in range(n_fireflies):
            if fitness_vals[j] < fitness_vals[i]:  # firefly j lebih baik (lebih terang)
                r = np.linalg.norm(fireflies[i] - fireflies[j])
                beta = beta0 * np.exp(-gamma * r**2)
                # Update posisi
                fireflies[i] = normalize(fireflies[i] + beta*(fireflies[j]-fireflies[i]) +
                                         alpha*(np.random.rand(n_assets)-0.5))
                fitness_vals[i] = fitness(fireflies[i], mean_return, cov_matrix, r_f)

# Hasil terbaik
best_idx = np.argmin(fitness_vals)
best_weights = fireflies[best_idx]
best_sharpe = -fitness_vals[best_idx]

# Hitung return & risiko dari portofolio terbaik
fa_return = np.dot(best_weights, mean_return)
fa_vol = np.sqrt(np.dot(best_weights.T, np.dot(cov_matrix, best_weights)))

print("Bobot Optimal Portofolio:")
for saham, w in zip(nama_perusahaan, best_weights):
    print(f"{saham}: {w:.4f}")

print("\nReturn Optimal:", round(fa_return, 4))
print("Risiko (Volatilitas) Optimal:", round(fa_vol, 4))
print("Sharpe Ratio Optimal:", round(best_sharpe, 4))
```

```python
# Visualisasi Efficient Frontier vs Solusi FA

n_portfolios = 5000
results = np.zeros((3, n_portfolios))

for i in range(n_portfolios):
    weights = np.random.dirichlet(np.ones(n_assets))
    port_return = np.dot(weights, mean_return)
    port_vol = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    sharpe = (port_return - r_f) / port_vol
    results[0,i] = port_vol
    results[1,i] = port_return
    results[2,i] = sharpe

plt.figure(figsize=(10,6))
plt.scatter(results[0,:], results[1,:], c=results[2,:], cmap='viridis', s=10)
plt.colorbar(label='Sharpe Ratio')

# Titik hasil Firefly Algorithm
plt.scatter(fa_vol, fa_return, color='red', marker='*', s=200, label='FA Optimal')

plt.title('Efficient Frontier vs Firefly Algorithm Optimal Portfolio')
plt.xlabel('Volatilitas (Risiko)')
plt.ylabel('Return Tahunan')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Hitung return dari data harga penutupan
data = close_data.drop(columns=["Tanggal"])
returns = data.pct_change().dropna()

# 2. Parameter pasar (prior)
cov = returns.cov()
n_assets = data.shape[1]

# Bobot pasar (equal weight kalau tidak ada market cap)
w_mkt = np.ones(n_assets) / n_assets
w_mkt = w_mkt.reshape(-1, 1)

# Risk aversion coefficient
delta = 2.5
tau = 0.05  # ketidakpastian

# Prior equilibrium excess return (pi)
pi = delta * cov.values.dot(w_mkt)
pi = pi.flatten()

# 3. Views investor
P = np.array([
    [1,  0,  0, -1,  0],   # BRI > BNI
    [0,  1,  0,  0, -1],   # BSI > Bank Mandiri
    [0,  0, -1,  1,  0],   # BTN < BNI
    [0,  0,  0, -1,  1],   # BNI < Bank Mandiri
    [-1, 0,  0,  0,  1]    # Bank Mandiri > BRI
])

# Ambil return terakhir
last_return = returns.iloc[-1].to_dict()

# Hitung Q otomatis sesuai view di P
views = [
    ("BRI", "BNI"),
    ("BSI", "Bank Mandiri"),
    ("BTN", "BNI"),
    ("BNI", "Bank Mandiri"),
    ("Bank Mandiri", "BRI")
]

factor = 2  # faktor pengali untuk agresivitas

Q = np.array([(last_return[long_asset] - last_return[short_asset]) * factor
              for long_asset, short_asset in views])

Omega = np.diag(np.full(len(Q), 0.01))
```

```python
import numpy as np
from scipy.optimize import minimize

# 4. Posterior expected return (Black-Litterman)
M_inv = np.linalg.inv(np.linalg.inv(tau * cov.values) + P.T @ np.linalg.inv(Omega) @ P)
adj_return = M_inv @ (np.linalg.inv(tau * cov.values) @ pi + P.T @ np.linalg.inv(Omega) @ Q)

# 5. Optimasi bobot portofolio dengan constraint w >= 0 dan sum(w) = 1
N = len(adj_return)
cov_matrix = cov.values

def objective(w):
    # Mean-variance: maximize return/risk -> minimize negative Sharpe-like
    port_return = w @ adj_return
    port_vol = np.sqrt(w.T @ cov_matrix @ w)
    return -port_return / port_vol  # negatif karena minimize

# Constraint: sum(w) = 1
constraints = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1})

# Bounds: 0 <= w_i <= 1
bounds = tuple((0, 1) for _ in range(N))

# Initial guess
w0 = np.array([1/N]*N)

# Optimasi
res = minimize(objective, w0, method='SLSQP', bounds=bounds, constraints=constraints)
w_bl = res.x
```

## **Metode Optimasi Multi-Objektif**


```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from scipy.optimize import minimize

# ================================
# 2. Expected return & variansi tiap saham
# ================================
returns = close_data.drop(columns=['Tanggal']).pct_change().dropna()

expected_return = returns.mean() * 52    # return tahunan
variance_each = returns.var() * 52     # variansi tahunan
cov_matrix = returns.cov() * 52          # kovariansi tahunan

print("=== Expected Return (Annual) ===")
print(expected_return)
print("\n=== Variance (Annual) ===")
print(variance_each)

# ================================
# 3. Fungsi portofolio
# ================================
def portfolio_performance(weights, mean_returns, cov_matrix):
    port_return = np.dot(weights, mean_returns)
    port_variance = np.dot(weights.T, np.dot(cov_matrix, weights))
    port_volatility = np.sqrt(port_variance)
    return port_return, port_variance, port_volatility

# Fungsi objektif sesuai teori: Min -w^T R + k * w^T Σ w
def objective(weights, mean_returns, cov_matrix, k=1):
    port_return = np.dot(weights, mean_returns)             # w^T R
    port_variance = np.dot(weights.T, np.dot(cov_matrix, weights))  # w^T Σ w
    return -port_return + k * port_variance

# ================================
# 4. Simulasi portofolio random (dengan Sharpe Ratio)
# ================================
num_assets = len(expected_return)
num_portfolios = 50000
risk_free_rate = 0.05  # risk free rate tahunan

results = np.zeros((4, num_portfolios))  # return, variance, volatility, sharpe
weights_record = []

for i in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)
    port_return, port_variance, port_volatility = portfolio_performance(weights, expected_return, cov_matrix)
    sharpe = (port_return - risk_free_rate) / port_volatility

    results[0, i] = port_return
    results[1, i] = port_variance
    results[2, i] = port_volatility
    results[3, i] = sharpe
    weights_record.append(weights)

# ================================
# 5. Optimasi Multi-Objektif
# ================================
k_value = 0.5  # bisa diubah untuk lihat pengaruh risk aversion

initial_weights = np.ones(num_assets) / num_assets
bounds = tuple((0, 1) for _ in range(num_assets))
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

opt_result = minimize(objective,
                      initial_weights,
                      args=(expected_return, cov_matrix, k_value),
                      method='SLSQP',
                      bounds=bounds,
                      constraints=constraints)

optimal_weights = opt_result.x
opt_return, opt_variance, opt_volatility = portfolio_performance(optimal_weights, expected_return, cov_matrix)
sharpe_ratio = (opt_return - risk_free_rate) / opt_volatility

# ================================
# 6. Hasil Portofolio Optimal + Sharpe Ratio
# ================================
portfolio_optimal = pd.DataFrame({
    "Perusahaan": expected_return.index,
    "Bobot Optimal": optimal_weights
}).sort_values(by="Bobot Optimal", ascending=False).reset_index(drop=True)

print("\n=== Portofolio Optimal ===")
print(portfolio_optimal)

print(f"\nExpected Annual Return: {opt_return:.2%}")
print(f"Portfolio Variance: {opt_variance:.4f}")
print(f"Portfolio Volatility: {opt_volatility:.2%}")
print(f"Sharpe Ratio: {sharpe_ratio:.4f}")

# ================================
# 7. Visualisasi berdasarkan Sharpe Ratio
# ================================
plt.figure(figsize=(10,6))
plt.scatter(results[1,:], results[0,:], c=results[3,:], cmap='viridis', marker='o', s=10)
plt.colorbar(label='Sharpe Ratio')

# Plot bintang portofolio optimal
plt.scatter(opt_variance, opt_return, c='red', marker='*', s=200, label='Optimal Portfolio')

# Tambahkan teks keterangan di dekat bintang
plt.annotate(
    f"Return: {opt_return:.2%}\nVolatility: {opt_volatility:.2%}\nVariance: {opt_variance:.4f}\nSharpe: {sharpe_ratio:.2f}",
    xy=(opt_variance, opt_return),
    xytext=(opt_variance+0.002, opt_return+0.005),
    arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6),
    fontsize=9,
    bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="white", alpha=0.7)
)

plt.xlabel('Variance (Risk)')
plt.ylabel('Expected Return')
plt.title('Portfolio Simulation (Sharpe Ratio Coloring)')
plt.legend()
plt.show()
```
